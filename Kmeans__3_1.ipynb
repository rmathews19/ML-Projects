```python

import pandas as pd #used for reading csv
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans
```

```python
X = pd.read_csv('Study_Hours_And_More.csv');
#X, y_true = make_blobs(n_samples = 300, centers = 4, cluster_std = 0.60, random_state = 0)
#plt.scatter(X[:, 0], X[:, 1], X[:,2], X[:,3], X[:,4], s = 50)

#REFERENCE
# https://stackoverflow.com/questions/43061768/plotting-multiple-scatter-plots-pandas

# sets the columns based on what the excel has
df = pd.DataFrame(X, columns=["Hours", "Tacos", "Electricity", "Mood", "Grade"])

# each line of ax is the combination of each column to have them under one scatter plot diagram
ax1 = df.plot(kind='scatter', x='Hours', y='Tacos', c='r')
ax2 = df.plot(kind='scatter', x='Electricity', y='Mood', c='g', ax=ax1)
# since there is an odd number of columns, the first and last are added together like a "cycle"
ax3 = df.plot(kind='scatter', x='Grade', y='Hours', c='b', ax=ax1)

#plt.scatter(X[:, 1], c="r")
#plt.scatter(X[:, 2], c="y")
#plt.scatter(X[:, 3], c="g")
#plt.scatter(X[:, 4], c="c")
```

```python
#Use K-means
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
#Plot results
ax1 = df.plot(kind='scatter', x='Hours', y='Tacos', c='r')
ax2 = df.plot(kind='scatter', x='Electricity', y='Mood', c='g', ax=ax1)
ax3 = df.plot(kind='scatter', x='Grade', y='Hours', c='b', ax=ax1)
centers = kmeans.cluster_centers_
# 3 diffrenet ones are made to accomandate having all columns being used
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.scatter(centers[:, 2], centers[:, 3], c='black', s=200, alpha=0.5);
plt.scatter(centers[:, 4], centers[:, 0], c='black', s=200, alpha=0.5);
```
